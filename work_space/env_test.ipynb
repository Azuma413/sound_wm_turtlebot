こちらは環境がうまくできているのかテストするためのもの。

```python
from envs.gym_env import MyEnv2 # カスタム環境の読み込み
import numpy as np
import matplotlib.pyplot as plt

env = MyEnv2() # 環境のインスタンス化
```

```python

obs, _ = env.reset() # 環境のリセット
# 画像の表示
print(obs.shape)
print(obs.dtype)
plt.imshow(obs)
print(obs)
```

```python
# エピソードの実行
done = False
#action = env.action_space.sample() # ランダムな行動の選択
action = np.array([0.5, -0.99])
print('action:', action) # 行動の表示
obs, reward, done, truncated, info = env.step(action) # 行動の実行
img = env.render()
print(img.shape)
print(img.dtype)
plt.imshow(img) # 画像の表示
plt.show()
print('reward:', reward) # 報酬の表示
print('info:', info) # infoの表示
print('done:', done) # エピソードの終了の表示
print('truncated:', truncated) # エピソードの終了の表示
```
