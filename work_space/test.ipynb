```python
import numpy as np
import cv2
import yaml
import os

yaml_path = os.getcwd() + "/map/" + str(327) + ".yaml"
with open(yaml_path, 'r') as file:
    map_data = yaml.safe_load(file)
    map_image = cv2.imread(os.getcwd() + "/map/" + map_data["image"], cv2.IMREAD_GRAYSCALE)
    resolution = map_data["resolution"] # 1pixelあたりのメートル数
    origin = np.array(map_data["origin"]) # mapの右下の隅のpose
```

```python
import matplotlib.pyplot as plt
plt.imshow(map_image, cmap='gray')
```
